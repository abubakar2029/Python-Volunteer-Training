```python
# You're given a person’s weight on Earth, and you need to calculate their weight on Mars.
# Gravity on Mars is about 38% of Earth’s gravity.

# Formula to calculate weight on Mars: mars_weight = earth_weight * 0.38

# Algorithm of the program
# input(), user sa weight ki value take krna ka lia
# earth_weight, hmm os ma person ka earth's weight store kr dan gan
# mars_weight, ya variable hmara result ko store kra ga
# print(), hmm os function sa output user ko show krean gan

earth_weight = float(input("enter your weight on earth"))
mars_weight = earth_weight * 0.38
print(mars_weight)
```
